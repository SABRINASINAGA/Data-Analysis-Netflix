```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('netflix_titles.csv')

data.head()
```

```python
data
```

```python
data.info()
```

```python
data["date_added"] = pd.to_datetime(data['date_added'])
data['year_added'] = data['date_added'].dt.year
data['month_added'] = data['date_added'].dt.month


data.head()
```

```python
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 15, 10
```

```python
d1 = data[data["type"] == "TV Show"]
d2 = data[data["type"] == "Movie"]

col = "year_added"

vc1 = d1[col].value_counts().reset_index()

vc1 = vc1.rename(columns = {col : "count", "index" : col})
vc1['percent'] = vc1['count'].apply(lambda x : 100*x/sum(vc1['count']))
vc1 = vc1.sort_values(col)

vc2 = d2[col].value_counts().reset_index()
vc2 = vc2.rename(columns = {col : "count", "index" : col})
vc2['percent'] = vc2['count'].apply(lambda x : 100*x/sum(vc2['count']))
vc2 = vc2.sort_values(col)

vc3 = data[col].value_counts().reset_index()
vc3 = vc3.rename(columns = {col : "count", "index" : col})
vc3['percent'] = vc3['count'].apply(lambda x : 100*x/sum(vc3['count']))
vc3 = vc3.sort_values(col)

print(vc3)
```

```python
sns.set_style("whitegrid")
plt.title("Content added over the years")
ax=sns.lineplot(x="year_added",y="count",data=vc3,label="Total",marker='o', markerfacecolor='blue', markersize=5)
ax=sns.lineplot(x="year_added",y="count",data=vc1,label="TV Shows",marker='o', markerfacecolor='orange', markersize=5)
ax=sns.lineplot(x="year_added",y="count",data=vc2,label="Movies",marker='o', markerfacecolor='green', markersize=5)
plt.subplot(ax)
plt.show()
```

```python
players_details, players_name = [], []
for each_player in data["type"]:
    if not(each_player in players_name):
        players_name = players_name + [each_player]
        players_details = players_details + [[each_player, 1]]
    else:
        for index in range(len(players_details)):
            if players_details[index][0] == each_player :
                players_details[index][1] = players_details[index][1]+1
channels_ = []
shows_available = []
for each in players_details:
    shows_available.append(each[0])
    channels_.append(each[1])
    #break
    print ('{} : {}'.format(*each))
print(shows_available[0])
```

```python
y = [channels_[0] ,channels_[1]]
x = [shows_available[0], shows_available[1]]
print(x)
print(y)
```

```python
plt.bar(x, y, color = ["red", "orange"])
plt.title("Which is Famous in Netflix\nMovie vs TV show ")
plt.xlabel("types of show available")
plt.ylabel("No.of show available")
plt.show()
```

```python
data_copy=data.copy()

data_copy.drop(['director','cast','date_added'],axis=1,inplace=True)

data_copy
```

```python
data_copy['country'].replace(np.NaN,'United States',inplace=True)

data_copy
```

```python
data_copy['rating']
```

```python
data_copy['rating'].replace(np.NaN,'TV-MA',inplace=True)

data_copy.head()
```

```python
data_copy.isnull().sum()
```

```python
data_copy.head()
```

```python
sns.countplot(data_copy['rating'])
sns.countplot(data_copy['rating']).set_xticklabels(sns.countplot(data_copy['rating']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('Rating');
```

```python
genre = data_copy["listed_in"].value_counts().tail(5)
genre
```

```python
plt.figure(figsize=(10,4))
genre.plot(kind="bar",color="blue")
plt.title("5 Least Popular Genres on Netflix",size=20);
```

```python
top_5_year = data_copy.groupby('release_year').count().sort_values('title',ascending=False).head(5)

top_5_year.reset_index(inplace=True)
top_5_year
```

```python
plt.bar(top_5_year.release_year,top_5_year.title);
```

```python
labels = ['Movie', 'TV show']
size = data_copy['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```
